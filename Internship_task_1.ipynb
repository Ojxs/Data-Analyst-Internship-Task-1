```python
import pandas as pd
```

```python
df = pd.read_csv('KaggleV2-May-2016.csv')
```

```python
# Print the current column names to see them clearly
print(df.columns)

# Rename the columns to a clean format
# The .str.lower() method converts all column names to lowercase.
# The .str.replace() method replaces spaces with underscores.
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Print the new column names to confirm the change
print(df.columns)
```

```python
# Convert the 'scheduledday' and 'appointmentday' columns to datetime objects
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])

# Print the info again to confirm the change in data types
df.info()
```

```python
# Convert both the 'scheduledday' and 'appointmentday' columns to include only the date part
df['scheduledday'] = df['scheduledday'].dt.date
df['appointmentday'] = df['appointmentday'].dt.date

# Check the data types again to confirm the change
df.info()
```

```python
# Convert the 'scheduledday' and 'appointmentday' columns to a string in 'dd-mm-yyyy' format
df['scheduledday'] = df['scheduledday'].dt.strftime('%d-%m-%Y')
df['appointmentday'] = df['appointmentday'].dt.strftime('%d-%m-%Y')

# Print the first few rows to confirm the change
df.head()
```
